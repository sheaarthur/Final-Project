```python
import random

def start():    #This is the function to start the game, and then includes what is said when the game is over
    greeting()    
    game=Pig()
    game.play()  #This is later defined in the Pig Class
    goodbye()
    
def greeting():
    print("Welcome to the game of Pig!")
    
def goodbye():
    print()
    print()
    print("Goodbye!")
```

```python
class Pig: #This class was created to define the general parts of the game
    
    def __init__(self,p1_player=True,p2_player=False): 
        self.die=Die()
        self.p1=Player('Player 1', p1_player)  #Where we created the 2 players
        self.p2=Player('Player 2', p2_player)
        
    def play(self):     #The function of playing the game
        while (self.p1.score<50 and self.p2.score<50):
            self.p1.move()      #While the total scores are less than 50, the game continues between the players
            if self.p1.score<50:
                self.p2.move()
        if (self.p1.score>self.p2.score):   #Whoevers score is >=50, they win the game
            print('Player 1 wins!')
        else:
            print('Player 2 wins!')
```

```python
class Player:
    
    def __init__(self,title,human_player=False):
        self.name=title
        self.is_human=human_player
        self.score=0   #Initialization of score
        self.die=Die(6)
    
    def move(self):    #This defines who's turn it is
        if self.is_human:
            self.human_move()
        else:
            self.computer_move()
            
    def human_move(self): #This is the users turn
        round_score=0    # This initializes the round score
        again='y'
        while again=='y':   #While the user continues to keep rolling
            self.die.roll()
            roll=self.die.face
            if roll==1:    #If roll==1, no round points are added to the total score and turn is over
                print(f'{self.name} rolled a 1.')
                round_score=0
                again='n'
            else:
                print(f'{self.name} rolled a {roll}.')  #The scores are added to the round score
                round_score=round_score+roll
                print(f"{self.name}'s round score is {round_score}.")
                again=input('Roll again? (y/n)')    #User is asked if tehy would like to roll again
        self.score+=round_score      #Round score is added to total score
        print(f"{self.name}'s turn is over.")
        print(f"{self.name}'s total score is {self.score}.\n\n")
        
    def computer_move(self):
        round_score=0  #Initialization of computer round score
        again='y'
        while again=='y':   #While the computer continues to keep rolling
            self.die.roll()
            roll=self.die.face
            if roll==1:    #If roll==1, no round points are added to the total score and turn is over
                print(f'{self.name} rolled a 1.')
                round_score=0
                again='n'
            else:
                print(f'{self.name} rolled a {roll}.')
                round_score=round_score+roll     #The scores are added to the round score
                if round_score<20:     #While the computers round score is <20, they will continue to roll
                    print(f'{self.name} will roll again.')
                else:      #Once the round score is >=20, the computer says no to rolling again
                    again='n'
        self.score+=round_score   #Round score is added to total score
        print('Turn is over.')
        print(f"{self.name}'s round score is {round_score}.")
        print(f"{self.name}'s total score is {self.score}.\n\n")
```

```python
class Die:   #This class was created for the simple task of rolling the dice
    
    def __init__(self,n=6): 
        self.sides=n
        self.roll()
        
    def roll(self):   #Chooses a random integer from [1,6] (including both 1 and 6) to represent dice being rolled
        self.face=int(random.random()*self.sides+1)
```
